```python
df = pd.read_csv("toloka_assignments.tsv", sep='\t')
```

```python
df["ASSIGNMENT:worker_id"].value_counts()
```

```python
pd.Series.tolist
```

```python
ratings_hum = []
ratings_gram = []

hum_sents = []
gram_sents = []

for frame in df.groupby('INPUT:headline'):
    sent_hum_ratings = frame[1]['OUTPUT:hum'].tolist()
    sent_gram_ratings = frame[1]['OUTPUT:gram'].tolist()
    
    if len(sent_hum_ratings) == 4:
        ratings_hum.append(sent_hum_ratings)
        hum_sents.append(frame[0])
    if len(sent_gram_ratings) == 4:
        ratings_gram.append(sent_gram_ratings)
        gram_sents.append(frame[0])
```

```python
ratings_hum
```

```python
pd.DataFrame(ratings_hum).corr(method='spearman')
```

```python
pd.DataFrame(ratings_hum).corr(method='kendall')
```

```python
pd.DataFrame(ratings_gram).corr(method='spearman')
```

```python
pd.DataFrame(ratings_gram).corr(method='kendall')
```
