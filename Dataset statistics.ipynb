```python
import pandas as pd
```

```python
df = pd.read_csv("unsupervised_news.csv")
```

```python
df
```

```python
df["source"].value_counts()
```

```python
df = df.drop(["content"], axis=1).dropna(subset=["title"])
```

```python
df["title"] = df["title"].apply(lambda x: x.lower().strip())
df_unique = df.drop_duplicates(subset=["title"])
```

```python
df_unique
```

```python
df_unique["source"].value_counts()
```

Попробуем выбросить AllTheNewsKaggle, снова выбросить дубликаты и посмотреть, сколько заголовков останется:

```python
df_unique1 = df[df['source']!="AllTheNewsKaggle"].drop_duplicates(subset=["title"])
```

```python
df_unique1["source"].value_counts()
```
